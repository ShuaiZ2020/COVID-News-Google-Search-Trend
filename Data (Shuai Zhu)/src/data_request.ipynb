```python
from pytrends.request import TrendReq
```

```python

pytrends = TrendReq(hl='en-US', tz=360)
kw_list = ['covid news']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='US-AZ', gprop='news')
```

```python
data = pytrends.interest_over_time()
data
```
